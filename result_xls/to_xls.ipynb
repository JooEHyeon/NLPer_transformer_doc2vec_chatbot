```
import os
import pickle
src_path = '/content/drive/My Drive/자연어처리,추천/NLPer/chatbot_pra/chatbot_basecode/pos_tagger_test'
folder_list = os.listdir(src_path)

data_dict = {}
for folder in folder_list[1:-1]:
    # print(folder)
    file_list = os.listdir(src_path+'/'+folder)
    # print(file_list)
    for file in file_list:
        with open(src_path+'/'+folder+'/'+ file, 'rb') as f:
            data_dict[file] = pickle.load(f)
        
```
