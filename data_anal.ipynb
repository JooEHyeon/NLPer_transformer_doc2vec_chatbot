```python
# '네'란 단어가 들어간 문장의 개수
ne_count = 0
for line in data.SENTENCE:
    word = line.split()
    for i in word:
        if i == '네':
            ne_count += 1
print(ne_count)
```

## SUB
 - 해당 값은 15826개의 문장중 1840개만이 이 값이 있다.
 - 170종류의 값이 올 수 있다.
 - 점원의 역질문과 해당 역질문에 대한 고객의 답변에 대한 주제 이때 해당 쌍은 같은 SUB값을 가진다.
 - 전체 문장중 점원의 역질문과 그에대한 답변의 개수가 총 1840개라는 의미는 아닌것이....
 - 점원의 역질문에 대한 고객의 대답 이후 고객의 역역질문이 이어진다면, NaN값이 채워진다. 즉, Q(고)AQ(점)A(고)구조의 뒤의 Q(점)A(고)에는 값이 있지만, Q(고)AQ(점)Q(고) or  Q(고)AQ(점)AQ(고)이런식의 두번 꼬여버린 경우 이후 뒤의 Q(고)부터의 문장들은 NaN값이 채워지는 듯..
 - 정리하면, 점원의 역질문과 이에 대한 고객의 답변에 해당하는 문장은 이 값을 가진다. 역질문의 의도? 그래서 MAIN의도가 아닌 SUB의도인듯..

```python
print("SUB 열의 값으로 올 수 있는 것 : ",data['SUB'].unique())
print("SUB 의 unique 개수 : ",data['SUB'].nunique())
print(data['SUB'].value_counts())
```

## MQ
 - 7570개의 문장만이 이 값을 지닌다.
 - 사용자(손님) 질문(Main Question)의 문장에만 있는 값. 고객의 질문 문장이 그대로 들어간다.

사용자(손님) 질문(Main Question), 메인 질문에 추가적으로 필요한 시스템(점원)의 서브 질문(Sub Question), 서브 질문에 대한 사용자(손님) 응답(User Answer), 시스템(점원) 최종 응답(System Answer)로 구분

```python
print("MQ 열의 값으로 올 수 있는 것 : ",data['MQ'].unique())
print("MQ 의 unique 개수 : ",data['MQ'].nunique())
print(data['MQ'].value_counts())
```

## SQ

```python
print("SQ 열의 값으로 올 수 있는 것 : ",data['SQ'].unique())
print("SQ 의 unique 개수 : ",data['SQ'].nunique())
print(data['SQ'].value_counts())
```

## SA
 - 6359개의 문장만이 이 값을 지닌다.

 - 점원의 답변에 해당하는 문장은 전부 이 값이 있다.
 - 8656(Q) - 7570(MQ) = 1086 = SQ : 점원의 질문

```python
print("SA 열의 값으로 올 수 있는 것 : ",data['SA'].unique())
print("SA 의 unique 개수 : ",data['SA'].nunique())
print(data['SA'].value_counts())
```
